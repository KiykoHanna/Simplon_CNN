```python
import os
import nltk



```

```python

nltk_data_dir = os.path.abspath('./.venv/nltk_data')

if nltk_data_dir not in nltk.data.path:
    nltk.data.path.append(nltk_data_dir)

nltk.download('punkt', download_dir=nltk_data_dir)
nltk.download('stopwords', download_dir=nltk_data_dir)
nltk.download('wordnet', download_dir=nltk_data_dir)
nltk.download('punkt_tab', download_dir=nltk_data_dir)
```

```python
from nltk.tokenize import word_tokenize

text = "L'intelligence artificielle c'est génial!"
tokens = word_tokenize(text)
tokens
```

```python
import re
def clean_text(text):
    text = re.sub(r"[^a-zA-Z0-9]+", " ", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text.lower()

clean_text = clean_text(text)
clean_text
```

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words('french'))
filtred = [w for w in tokens if w.lower() not in stop_words and w.isalnum()]
filtred
```

```python
tokens = word_tokenize(clean_text)
tokens
```

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

stems = [stemmer.stem(m) for m in filtred]
stems
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

lemms = [lemmatizer.lemmatize(m) for m in filtred]
lemms
```

```python
mots = ["cars", "careful", "carefully", "caring", "care"]
lemmated = [lemmatizer.lemmatize(m) for m in mots]
lemmated
```
