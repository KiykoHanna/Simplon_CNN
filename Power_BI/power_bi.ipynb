```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
df = sns.load_dataset('iris')
df = sns.load_dataset('iris')
df.head()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5))
sns.scatterplot(data=df, x="sepal_length", y="sepal_width", hue="species", ax=axes[0])
sns.scatterplot(data=df, x="petal_length", y="petal_width", hue="species", ax=axes[1])
```

```python

sns.jointplot(data=df, x="petal_length", y="petal_width",
kind="scatter", hue="species", palette="Set2")
plt.show()
```

```python
list_kind = ["kde", "hist"]
for kind in list_kind:
    sns.jointplot(
        data=df,
        x="petal_length",
        y="petal_width",
        kind=kind,
        hue="species",
        palette="Set2"
    )
```

```python
sns.jointplot(
        data=df,
        x="petal_length",
        y="petal_width",
        kind="hex",
        palette="Set2"
    )
```

```python
df_feat = df.drop(columns="species")
corr_matrix = df_feat.corr()
sns.heatmap(corr_matrix, annot=True, cmap=sns.cubehelix_palette(as_cmap=True))
plt.show()
```

```python
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
sns.heatmap(corr_matrix, annot=True, cmap=sns.cubehelix_palette(as_cmap=True), mask=mask)
plt.show()
```
